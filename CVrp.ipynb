```python
#importing 
import pandas as pd
import os
import csv
import matplotlib.pyplot as plt

```

```python
#reading CSV
corona_virus = pd.read_csv("2019_nCoV_data.csv")
corona_virus
```

```python
without_china = corona_virus.loc[corona_virus["Country"] !="Mainland China"]
countries_cv = without_china.groupby("Country").sum()
countries_updated = countries_cv[['Deaths']]
```

```python
countries_updated.plot(kind="bar", figsize=(10,5))
plt.title("Deaths by Country Excluding China")
plt.show()
plt.tight_layout()
```

```python
countries_updated.plot.pie(y='Deaths', figsize=(15, 15))
plt.title("Deaths by Province/State")
plt.show()
plt.tight_layout()
```

```python
corona_deaths = corona_virus.groupby("Country").sum()
corona_deaths
```

```python
country_deaths_filtered = corona_deaths[['Deaths']]
country_deaths_filtered
```

```python
country_deaths.plot.pie(y='Deaths', figsize=(15, 15))
```

```python
#Starting Country Analysis
#Dataframe without Hubei
without_hubei = corona_virus.loc[corona_virus["Province/State"] !="Hubei"]
without_hubei.head()
```

```python
#grouping by Province/State
country_cv = without_hubei.groupby("Province/State").sum()
#Showing deaths per province/state
country_cv["Deaths"]
#Showing deaths per province/state > 0
deaths = country_cv[country_cv["Deaths"]>0]
only_deaths = deaths[['Deaths']]
only_deaths
```

```python
#plotting bar graph

only_deaths.plot(kind="bar", figsize=(10,5))
plt.title("Deaths by Province/State")
plt.show()
plt.tight_layout()
```

```python
#plotting pie chart

only_deaths.plot.pie(y='Deaths', figsize=(15, 15))
plt.title("Deaths by Province/State")
plt.show()
plt.tight_layout()
```

```python
#starting time analysis
#grouping by date

date_cv = corona_virus.groupby("Date").sum()

#Taking serial no. out
no_sno = date_cv[['Confirmed','Recovered','Deaths']]
no_sno

```

```python
#Plotting Line Chart

plt.figure(figsize=(20,10))
no_sno.plot()
plt.title("Confirmed vs. Recovered vs. Deaths")
plt.xlabel("Date")
plt.ylabel("Number of Cases")
plt.xticks(rotation=45)
plt.show
```

```python
#Plotting Second Line Chart Without Confirmed Cases

no_confirmed = date_cv[['Recovered','Deaths']]
no_confirmed
no_confirmed.plot()
plt.title("Recovered vs. Deaths")
plt.xlabel("Date")
plt.ylabel("Number of Cases")
plt.xticks(rotation=45)
plt.show
```

```python
#Plotting Third Line Chart Without Recovered Cases

no_recovered = date_cv[['Deaths']]
no_recovered
no_recovered.plot()
plt.title("Deaths")
plt.xlabel("Date")
plt.ylabel("Number of Cases")
plt.xticks(rotation=45)
plt.show
```
